```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

from IPython.display import Markdown, display
```

```python
kmi30 = pd.read_csv("seasonality/Karachi Meezan 30 Historical Data.csv")

kmi30['Date'] = pd.to_datetime(kmi30['Date'])

kmi30['Open'] = kmi30['Open'].map(lambda x: str(x).replace(",", ""))
kmi30["LogReturns"] = np.log(kmi30['Open'].astype("float")).diff()
kmi30 = kmi30.drop(0)

kmi30.head()
```

```python
kse30 = pd.read_csv("seasonality/Karachi 30 Historical Data.csv")

kse30['Date'] = pd.to_datetime(kse30['Date'])

kse30['Open'] = kse30['Open'].map(lambda x: str(x).replace(",", ""))
kse30["LogReturns"] = np.log(kse30['Open'].astype("float")).diff()
kse30 = kse30.drop(0)

kse30.head()
```

```python
kse100 = pd.read_csv("seasonality/Karachi 100 Historical Data.csv")

kse100['Date'] = pd.to_datetime(kse100['Date'], format="%b %y")

kse100['Open'] = kse100['Open'].map(lambda x: str(x).replace(",", ""))
kse100["LogReturns"] = np.log(kse100['Open'].astype("float")).diff()
kse100 = kse100.drop(0)

kse100.head()
```

```python
def monthOfYear (df):
    df['Jan'] = (df['Date'].dt.month == 1).astype("int")
    df['Feb'] = (df['Date'].dt.month == 2).astype("int")
    df['Mar'] = (df['Date'].dt.month == 3).astype("int")
    df['Apr'] = (df['Date'].dt.month == 4).astype("int")
    df['May'] = (df['Date'].dt.month == 5).astype("int")
    df['Jun'] = (df['Date'].dt.month == 6).astype("int")
    df['Jul'] = (df['Date'].dt.month == 7).astype("int")
    df['Aug'] = (df['Date'].dt.month == 8).astype("int")
    df['Sep'] = (df['Date'].dt.month == 9).astype("int")
    df['Oct'] = (df['Date'].dt.month == 10).astype("int")
    df['Nov'] = (df['Date'].dt.month == 11).astype("int")
    #df['Dec'] = (df['Date'].dt.month == 12).astype("int")
    
    return df
```

```python
def quarterOfTheYear (df):
    df['Q1'] = (df['Date'].dt.month <= 3).astype("int")
    df['Q2'] = ((df['Date'].dt.month > 3) & (df['Date'].dt.month <= 6)).astype("int")
    df['Q3'] = ((df['Date'].dt.month <= 9) & (df['Date'].dt.month > 6)).astype("int")
    #df['Q4'] = (df['Date'].dt.month > 9).astype("int")
    
    return df
```

```python
def winter (df):
    df['Winter'] = ((df['Date'].dt.month > 8) & (df['Date'].dt.month < 2)).astype("int")
    
    return df
```

```python
def summer (df):
    df['Summer'] = ((df['Date'].dt.month > 4) & (df['Date'].dt.month < 8)).astype("int")
    
    return df
```

```python
def dayOfTheWeek (df):
    df['Mon'] = (df['Date'].dt.day == 1).astype("int")
    df['Tue'] = (df['Date'].dt.day == 2).astype("int")
    df['Wed'] = (df['Date'].dt.day == 3).astype("int")
    df['Thu'] = (df['Date'].dt.day == 4).astype("int")
    #df['Fri'] = (df['Date'].dt.day == 5).astype("int")
    #df['Sat'] = (df['Date'].dt.day == 6).astype("int")
    
    return df
```

```python
def weekOfYear (df):
    return pd.get_dummies(df['Date'].dt.weekofyear, prefix="Week")
```

```python
def militaryRule (df):
    df['MilitaryRule'] = ( ( (df['Date'].dt.year > 1978) & (df['Date'].dt.year < 1999) ) | 
                          ( (df['Date'].dt.year > 2000) & (df['Date'].dt.year < 2009) ) ).astype("int")
    
    return df
```

```python
def ramzan (df):
    df['Ramzan'] = ( df['Date'].between(pd.datetime(2018, 5, 15), pd.datetime(2018, 6, 16)) | 
                    df['Date'].between(pd.datetime(2017, 5, 27), pd.datetime(2017, 6, 26)) | 
                    df['Date'].between(pd.datetime(2016, 6, 5), pd.datetime(2016, 7, 6)) | 
                    df['Date'].between(pd.datetime(2015, 6, 17), pd.datetime(2017, 7, 16)) | 
                    df['Date'].between(pd.datetime(2014, 6, 28), pd.datetime(2014, 7, 29)) | 
                    df['Date'].between(pd.datetime(2013, 7, 28), pd.datetime(2013, 8, 29)) | 
                    df['Date'].between(pd.datetime(2012, 7, 19), pd.datetime(2012, 8, 18)) | 
                    df['Date'].between(pd.datetime(2011, 7, 31), pd.datetime(2011, 8, 19)) | 
                    df['Date'].between(pd.datetime(2010, 8, 10), pd.datetime(2010, 9, 9)) | 
                    df['Date'].between(pd.datetime(2009, 8, 21), pd.datetime(2009, 9, 20)) | 
                    df['Date'].between(pd.datetime(2008, 9, 1), pd.datetime(2008, 9, 30)) | 
                    df['Date'].between(pd.datetime(2007, 9, 12), pd.datetime(2007, 10, 12)) 
                    ).astype("int")
    
    return df
```

```python
def holiday (df):
    df['DayBeforeHoliday'] = (((df['Date'].dt.day == 31) & (df['Date'].dt.month == 12)) |
                    ((df['Date'].dt.day == 4) & (df['Date'].dt.month == 2)) |
                    ((df['Date'].dt.day == 22) & (df['Date'].dt.month == 3)) |
                    ((df['Date'].dt.day == 30) & (df['Date'].dt.month == 4)) |
                    ((df['Date'].dt.day == 30) & (df['Date'].dt.month == 6)) |
                    ((df['Date'].dt.day == 13) & (df['Date'].dt.month == 8)) |
                    ((df['Date'].dt.day == 8) & (df['Date'].dt.month == 11)) |
                    ((df['Date'].dt.day == 24) & (df['Date'].dt.month == 12))
                    #(df['Date'].dt.day == 5)
                    ).astype("int")
    df['DayAfterHoliday'] = (((df['Date'].dt.day == 2) & (df['Date'].dt.month == 1)) |
                    ((df['Date'].dt.day == 6) & (df['Date'].dt.month == 2)) |
                    ((df['Date'].dt.day == 24) & (df['Date'].dt.month == 3)) |
                    ((df['Date'].dt.day == 2) & (df['Date'].dt.month == 5)) |
                    ((df['Date'].dt.day == 2) & (df['Date'].dt.month == 7)) |
                    ((df['Date'].dt.day == 15) & (df['Date'].dt.month == 8)) |
                    ((df['Date'].dt.day == 9) & (df['Date'].dt.month == 11)) |
                    ((df['Date'].dt.day == 26) & (df['Date'].dt.month == 12))
                    #(df['Date'].dt.day == 1)
                    ).astype("int")
    return df
```

```python
def eidUlAzha (df):
    df['BeforeEid'] = (((df['Date'].dt.day == 1) & (df['Date'].dt.month == 9) & (df['Date'].dt.year == 2017)) | 
                       ((df['Date'].dt.day == 12) & (df['Date'].dt.month == 9) & (df['Date'].dt.year == 2016)) | 
                       ((df['Date'].dt.day == 24) & (df['Date'].dt.month == 9) & (df['Date'].dt.year == 2015)) |
                       ((df['Date'].dt.day == 5) & (df['Date'].dt.month == 10) & (df['Date'].dt.year == 2014)) | 
                       ((df['Date'].dt.day == 15) & (df['Date'].dt.month == 10) & (df['Date'].dt.year == 2013)) | 
                       ((df['Date'].dt.day == 26) & (df['Date'].dt.month == 10) & (df['Date'].dt.year == 2012)) | 
                       ((df['Date'].dt.day == 6) & (df['Date'].dt.month == 11) & (df['Date'].dt.year == 2011)) | 
                       ((df['Date'].dt.day == 16) & (df['Date'].dt.month == 11) & (df['Date'].dt.year == 2010)) | 
                       ((df['Date'].dt.day == 27) & (df['Date'].dt.month == 11) & (df['Date'].dt.year == 2009)) | 
                       ((df['Date'].dt.day == 7) & (df['Date'].dt.month == 12) & (df['Date'].dt.year == 2008)) | 
                       ((df['Date'].dt.day == 20) & (df['Date'].dt.month == 12) & (df['Date'].dt.year == 2007))
                      ).astype("int")
    return df
```

```python
def eidUlFitr (df):
    df['BeforeEid'] = (((df['Date'].dt.day == 15) & (df['Date'].dt.month == 7) & (df['Date'].dt.year == 2018)) | 
                       ((df['Date'].dt.day == 25) & (df['Date'].dt.month == 7) & (df['Date'].dt.year == 2017)) | 
                       ((df['Date'].dt.day == 6) & (df['Date'].dt.month == 7) & (df['Date'].dt.year == 2016)) | 
                       ((df['Date'].dt.day == 17) & (df['Date'].dt.month == 7) & (df['Date'].dt.year == 2015)) |
                       ((df['Date'].dt.day == 28) & (df['Date'].dt.month == 7) & (df['Date'].dt.year == 2014)) | 
                       ((df['Date'].dt.day == 7) & (df['Date'].dt.month == 8) & (df['Date'].dt.year == 2013)) | 
                       ((df['Date'].dt.day == 18) & (df['Date'].dt.month == 8) & (df['Date'].dt.year == 2012)) | 
                       ((df['Date'].dt.day == 30) & (df['Date'].dt.month == 8) & (df['Date'].dt.year == 2011)) | 
                       ((df['Date'].dt.day == 9) & (df['Date'].dt.month == 9) & (df['Date'].dt.year == 2010)) | 
                       ((df['Date'].dt.day == 20) & (df['Date'].dt.month == 9) & (df['Date'].dt.year == 2009)) | 
                       ((df['Date'].dt.day == 1) & (df['Date'].dt.month == 10) & (df['Date'].dt.year == 2008)) | 
                       ((df['Date'].dt.day == 12) & (df['Date'].dt.month == 10) & (df['Date'].dt.year == 2007))
                      ).astype("int")
    return df
```
